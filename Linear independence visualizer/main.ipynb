```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Taking vector as a input from the user

vector_1 = list(map(int, input("Enter vector one coordinates: ").split()))

vector_2 = list(map(int, input("Enter vector two coordinates: ").split()))

print(vector_1)
print(vector_2)
```

```python
plt.quiver(0, 0, vector_1[0], vector_1[1], angles = 'xy', scale_units = 'xy', scale = 1, color = "blue", label = "Vector 1")
plt.quiver(0, 0, vector_2[0], vector_2[1], angles = 'xy', scale_units = 'xy', scale = 1, color = "green", label = "Vector 2")

max_x = max(abs(vector_1[0]), abs(vector_2[0]))
max_y = max(abs(vector_1[1]), abs(vector_2[1]))

padding = 2

plt.xlim(-(max_x + padding), max_x + padding)
plt.ylim(-(max_y + padding), max_y + padding)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)



plt.grid(True)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(111, projection="3d") #Syntax --> fig.add_subplot(rows, columns, index)

'''
I wrote projections = 3d so that it create a 3d axes
'''
v1 = [2, 3, 4]
v2 = [-4, -2, 2]

ax.quiver(
    0, 0, 0,
    v1[0], v1[1], v1[2],
    color="blue"
)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color = "yellow")

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)
```
